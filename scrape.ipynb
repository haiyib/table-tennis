```python
import pandas as pd
import requests
from io import StringIO
import warnings
warnings.filterwarnings('ignore')

url = "https://en.wikipedia.org/wiki/ITTF_World_Ranking"
response = requests.get(url, verify=False)

# Explicitly specify lxml flavor
tables = pd.read_html(StringIO(response.text), flavor='lxml')

print(f"Found {len(tables)} tables")
for i, table in enumerate(tables):
    print(f"\n--- Table {i} ---")
    print(table.head(3))
```

```python
import requests
import warnings
from bs4 import BeautifulSoup
import pandas as pd

warnings.filterwarnings('ignore')

url = "https://en.wikipedia.org/wiki/ITTF_World_Ranking"
response = requests.get(url, verify=False, headers={"User-Agent": "Mozilla/5.0"})
soup = BeautifulSoup(response.text, 'lxml')

# Find all wikitables
all_tables = soup.find_all('table', class_='wikitable')
print(f"Found {len(all_tables)} wikitables")

# Parse each one into a DataFrame
for i, table in enumerate(all_tables):
    try:
        df = pd.read_html(str(table), flavor='lxml')[0]
        print(f"\n--- Table {i} ---")
        print(f"Shape: {df.shape}")
        print(df.head(5))
    except Exception as e:
        print(f"Table {i} failed: {e}")
```
