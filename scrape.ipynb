```python
"""
STEP 4: Visualize China's dominance using plotnine (ggplot for Python)
Install first:
    pip install plotnine
"""

from plotnine import (
    ggplot, aes, geom_bar, geom_line, geom_point, geom_hline,
    facet_wrap, scale_fill_manual, scale_color_manual,
    theme, theme_minimal, element_text, element_blank,
    labs, ylim
)
import pandas as pd

# --- Prep: count China vs non-China in top 5 per year/gender ---
df['Is_China'] = df['Country'].apply(lambda x: 'China' if x == 'China' else 'Others')

china_summary = (
    df.groupby(['Year', 'Gender', 'Is_China'])
    .size()
    .reset_index(name='Count')
)

# --- Plot: stacked bar chart ---
p = (
    ggplot(china_summary, aes(x='Year', y='Count', fill='Is_China'))
    + geom_bar(stat='identity', position='stack', width=0.7)
    + facet_wrap('~Gender', ncol=1)
    + scale_fill_manual(
        values={'China': '#DE2910', 'Others': '#cccccc'},
        name=''
    )
    + geom_hline(yintercept=5, linetype='dashed', color='black', size=0.4)
    + labs(
        title="China's Dominance in ITTF World Rankings Top 5 (2005–2025)",
        subtitle="Red = Chinese players | Dashed line = all 5 spots",
        x='Year',
        y='Players in Top 5',
        caption='Source: ITTF World Rankings'
    )
    + ylim(0, 5.5)
    + theme_minimal()
    + theme(
        plot_title   = element_text(size=13, weight='bold'),
        plot_subtitle= element_text(size=10, color='#555555'),
        axis_text_x  = element_text(angle=45, hjust=1, size=8),
        strip_text   = element_text(size=11, weight='bold'),
        legend_position='top',
        figure_size  = (10, 7)
    )
)

print(p)
p.save('ittf_china_dominance.pdf', dpi=150)
print("✅ Saved → ittf_china_dominance.pdf")
```
